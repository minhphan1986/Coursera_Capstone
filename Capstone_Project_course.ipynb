```python
# read data from wiki to dataframe
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M",skiprows=0)[0]
df.columns.values[0]='PostalCode'
df.head()
df.shape
```

```python
# Delete row with 'Not assigned' value in Borough
indexNames = df[ df['Borough'] == 'Not assigned' ].index
df.drop(indexNames , inplace=True)
df
```

```python
df['Neighbourhood']=df['Neighbourhood'].replace('Not assigned', df['Borough'])
df
```

```python
# Group the PostalCode

df2=df.groupby(['PostalCode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
```

```python
# FOR MARKING - df2 is the final output
df2.shape
```

```python
# Read the geographical coordinates from csv file
geo = pd.read_csv('http://cocl.us/Geospatial_data')
geo.columns.values[0]='PostalCode'
geo.head()
```

```python
# Merge the two dataframes

df3 = pd.merge(df2, geo, on='PostalCode')
df3
```
